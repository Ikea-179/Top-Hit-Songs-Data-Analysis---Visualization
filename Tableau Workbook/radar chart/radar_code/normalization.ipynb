```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/alan/Desktop/hottest_50_1960_2022_radar.csv')
df.columns
```

```python
radar_feature = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'liveness', 'valence']
```

```python
df_for_normalize = df[radar_feature]
normalized_df = (df_for_normalize - df_for_normalize.min()) / (df_for_normalize.max() - df_for_normalize.min())
```

```python
normalized_df.head()
```

```python
normalized_df['year'] = df['year']
y1960 = normalized_df[normalized_df['year'] == 1960]
y1960_mean = y1960.mean()
year_mean = pd.DataFrame(data = [], columns = dict(y1960_mean).keys())

for year in normalized_df.groupby('year'):
        year_mean.loc['new'] = list(year[1].mean())
        year_mean = year_mean.reset_index(drop = True)
year_mean = year_mean.round(2)
year_mean['year'] = year_mean['year'].astype(int)
```

```python
year_mean.head()
```

```python
year_mean.to_csv('year_mean.csv', index = None)
```

```python
decade_to_convert = year_mean
decade_to_convert['era'] = year_mean['year'].apply(lambda x: int(str(x)[0:3]))
```

```python
decade_to_convert.head()
```

```python
d1960 = decade_to_convert[decade_to_convert['era'] == 196]
d1960_mean = d1960.mean()
decade_mean = pd.DataFrame(data = [], columns = dict(d1960_mean).keys())

for decade in decade_to_convert.groupby('era'):
        decade_mean.loc['new'] = list(decade[1].mean())
        decade_mean = decade_mean.reset_index(drop = True)
decade_mean = decade_mean.round(2)
```

```python
decade_mean.drop('year', axis = 1, inplace = True)
```

```python
decade_mean['era'] = decade_mean['era'] * 10
```

```python
decade_mean['era'] = decade_mean['era'].astype(int) 
decade_mean['era'] = decade_mean['era'].astype(str) + 's'
decade_mean
```

```python
decade_mean.to_csv('decade_mean.csv', index = None)
```

```python
df_since_2010 = df[df['year'] >= 2010]
```

```python
df_since_2010 = df_since_2010.sort_values('weeks', ascending = False)
```

```python
df_since_2010.head()
```

```python
top_20 = df_since_2010.iloc[0:20]
bottom_20 = df_since_2010.iloc[-20:]
top_20 = top_20[radar_feature]
bottom_20 = bottom_20[radar_feature]
top_20_normalized = (top_20 - top_20.min()) / (top_20.max() - top_20.min())
bottom_20_normalized = (bottom_20 - bottom_20.min()) / (bottom_20.max() - bottom_20.min())
```

```python
top_20_normalized_mean = top_20_normalized.mean()
top_20_normalized_mean
```

```python
bottom_20_normalized_mean = bottom_20_normalized.mean()
bottom_20_normalized_mean
```

```python

comparison_mean = pd.DataFrame(data = [], columns = dict(top_20_normalized_mean).keys())

for group in [top_20_normalized, bottom_20_normalized]:
        comparison_mean.loc['new'] = list(group.mean())
        comparison_mean = comparison_mean.reset_index(drop = True)
comparison_mean = comparison_mean.round(2)
```
