```python
genre_list = ['country', 'rock', 'metal', 'r&b', 'rap', 'hip hop', 'soul', 'pop']
```

```python
import pandas as pd
import ast
```

```python
file = pd.read_csv('/Users/Desktop/hottest_50_1960_2022_new.csv')
file.head()
```

```python
genre_encoding = []
for index in range(0, len(file)):
# for index in range(135, 140):

    this_genre_encoding_dict = {'country': 0, 'rock': 0, 'r&b': 0, 'rap': 0, 'hip hop': 0, 'soul': 0, 'pop': 0}
    this_genre = ast.literal_eval(file.iloc[index]['genres'])
    this_genre = [n.strip() for n in this_genre]
    for item in this_genre:
        for genre in genre_list:
            if genre == 'metal':
                if genre in item.lower():
                    this_genre_encoding_dict['rock'] += 1
                    break
            else:
                if genre in item.lower():
                    this_genre_encoding_dict[genre] += 1
                    break
    
#     print(this_genre_encoding_dict)
                
    this_genre_encoding_dict = {k: v for k, v in sorted(this_genre_encoding_dict.items(), key = lambda item: item[1], reverse = True)}
    first_key = list(this_genre_encoding_dict.keys())[0]
    
#     print(this_genre_encoding_dict)
    if first_key == 'rap' or first_key == 'hip hop':
        genre_encoding.append('rap/hip hop')
    else:
        if this_genre_encoding_dict[first_key] > this_genre_encoding_dict['rap'] + this_genre_encoding_dict['hip hop']:
            genre_encoding.append(first_key)
        elif this_genre_encoding_dict[first_key] < this_genre_encoding_dict['rap'] + this_genre_encoding_dict['hip hop']:
            genre_encoding.append('rap/hip hop')
        else:
            genre_encoding.append(None)
#     print(genre_encoding)
#     print()
                
```

```python
file['genre_encoding'] = genre_encoding
```

```python
file.to_csv('/Users/Desktop/hottest_50_1960_2022_encoding.csv', index = None, encoding = 'utf-8-sig')
```
